```python
con=psycopg2.connect(user=sqluser,
                     password=pw,
                     host='/var/run/postgresql/',
                     dbname=targetdb)
con.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT)
cursor = con.cursor()
```

```python
package_location = '/home/ubuntu/software/munge_2.0/'
df1 = pd.read_csv(package_location + 'remap_ethnicity.csv')
df2 = pd.read_csv(package_location + 'remap_gender_GenderM.csv')
df3 = pd.read_csv(package_location + 'remap_language.csv')
df4 = pd.read_csv(package_location + 'remap_marital_status.csv')
df5 = pd.read_csv(package_location + 'remap_religion.csv')

df1 = df1[['Category','Remap0']]
df2 = df2[['Category','Remap0']]
df3 = df3[['Category','Remap0']]
df4 = df4[['Category','Remap0']]
df5 = df5[['Category','Remap0']]

ethn_dict = df1.set_index('Category').T.to_dict('list')
gender_dict = df2.set_index('Category').T.to_dict('list')
language_dict = df3.set_index('Category').T.to_dict('list')
marit_dict = df4.set_index('Category').T.to_dict('list')
relig_dict = df5.set_index('Category').T.to_dict('list')

df = pd.read_sql('select * from context;',con)
df = df.replace({'ethnicity':ethn_dict})
df = df.replace({'gender':gender_dict})
df = df.replace({'language':language_dict})
df = df.replace({'marital_status':marit_dict})
df = df.replace({'religion':relig_dict})
```
