```python
import dash
import dash_core_components as dcc
import dash_html_components as html

import datetime as dt
import flask
import os
import pandas as pd
import time
import plotly.graph_objs as go
from dash.dependencies import Input, Output, State

from scipy import stats


app = dash.Dash()
#real data
df = pd.read_csv('/home/eric/dash/contextplus.csv')
#synthetic data
df2 = pd.read_csv('/home/eric/dash/autoencoded_cohort_with_nans.csv')

mapp = pd.read_csv('/home/eric/dash/nitem_map.csv')

tcols = [col for col in df if col.startswith('tot')]
avcols = [col for col in df if col.startswith('avr')]

mapp['n_concat']= "_"+mapp['nid_class1'].astype("str")+"_"+mapp['nid_class2'].astype("str")
umapp = mapp.drop_duplicates('class2')
umapp = umapp[['nid_table','class2','n_concat']]

ce_umapp = umapp[umapp['nid_table']==1]
ce_umapp['old'] = 'avrg'+ce_umapp['n_concat']
ce_umapp['label'] = 'Average Value: '+ce_umapp['class2']
in_umapp = umapp[umapp['nid_table']==2]
in_umapp['old'] = 'tot'+in_umapp['n_concat']+"_in"
in_umapp['label'] = 'Total Input: '+in_umapp['class2']
out_umapp = umapp[umapp['nid_table']==3]
out_umapp['old'] = 'tot'+out_umapp['n_concat']+"_out"
out_umapp['label'] = 'Total Output: '+out_umapp['class2']

newdf = ce_umapp[['old','label']]
newdf = newdf.append(in_umapp[['old','label']])
newdf = newdf.append(out_umapp[['old','label']])


tempcols = newdf.set_index('old').T.to_dict('index')
newcols = tempcols['label']
tempcols2 = {'age_at_death' : 'Age at Death',
          'age_at_adm':'Age',
          'los_total':'Length of Stay'}
newcols.update(tempcols2)

df.rename(columns=newcols, inplace=True)
df2.rename(columns=newcols, inplace=True)
cols = list(newcols.values())

mrkdown_text = '''
##### What am I looking for here?  
**tl;dr** Look for similar distributions (histograms and scatterplots) between 
the real data and the synthetic data.  

This cohort contains patients who've undergone any cardiac procedure and
have stayed in critical care units in the Boston Greater Area. The size 
of this subset is ~8000 patients with ~4000 features including demographic 
(such as gender, insurance, religion, age, ethnicity, etc.) clinical (such 
as diagnoses, procedures, medication, etc.) and physiological (including blood
tests, fluids entering/leaving the body, transfusions, nutrients, etc.)  
From this, our algorithm learned the distribution and relationship between
all 4000 features, to create a dataset populated with synthetic patients. 
This application showcases the faithful recreation of individual variables 
(histograms) as well as multivariables (scatterplots). 
  
  
*Data source*:    
MIMIC-III, a freely accessible critical care database. Johnson AEW,  
Pollard TJ, Shen L, Lehman L, Feng M, Ghassemi M, Moody B, Szolovits P,  
Celi LA, and Mark RG. Scientific Data (2016). DOI: 10.1038/sdata.2016.35.  
Available from: [here](http://www.nature.com/articles/sdata201635)

'''

app.layout = html.Div([
    html.Div([
        html.H2('COMPARISON BETWEEN REAL AND SYNTHETIC',
                style={'display': 'inline',
                       'float': 'left',
                       'font-size': '2.65em',
                       'margin-left': '7px',
                       'font-weight': 'bolder',
                       'font-family': 'Product Sans',
                       'color': "rgba(117, 117, 117, 0.95)",
                       'margin-top': '20px',
                       'margin-bottom': '0'
                       }),
        html.H4('To get started: select 1 variable to compare histograms, or select 2 variables to compare scatterplots',
                style={'display': 'inline',
                       'float': 'left',
                       'font-size': '1.2em',
                       'margin-left': '7px',
                       'font-weight': 'bolder',
                       'font-family': 'Product Sans',
                       'color': "rgba(85, 242, 152, 0.75)",
                       'margin-top': '20px',
                       'margin-bottom': '0'
                       })
                
#         , html.Img(src=image_filename,
#                 style = {
#                     'height':'100px',
#                     'float': 'right'
#                 })
    ]),
    dcc.Dropdown(
        id='dropdown1',
        options=[{'label': t, 'value': t} 
                 for t in cols],
        placeholder="select up to two variables to generate plots"
        ,multi=True
                ),    
    html.Div(id='debugger'),
    html.Div([
    html.Div(id='graph1', className="six columns"),
    html.Div(id='graph2', className="six columns")
    ],className="row"),
    
    dcc.Markdown(mrkdown_text)

], className="container")



def graph_histogram1(somedata, title, dataframe,gid):
        return dcc.Graph(
                figure=go.Figure(
                    data=[go.Histogram(x=dataframe[somedata].values.tolist(),
                                      marker=dict(
                                          color = 'rgb(0,0,153)')
                                      )
                         ],
                    layout=go.Layout(
                        title=title,
                        margin=go.Margin(l=40, r=0, t=40, b=30)
                    )
                ),
                style={'height': 300},
            id='grapha'
            )

def graph_histogram2(somedata, title, dataframe,gid):
        return dcc.Graph(
                figure=go.Figure(
                    data=[go.Histogram(x=dataframe[somedata].values.tolist(),
                                      marker=dict(
                                          color = 'rgb(0,0,153)')
                                      )
                         ],
                    layout=go.Layout(
                        title=title,
                        margin=go.Margin(l=40, r=0, t=40, b=30)
                    )
                ),
                style={'height': 300},
            id='graphb'
            )
    
def graph_scatter1(somedata1, somedata2, title, dataframe,gid):
        return dcc.Graph(
                figure=go.Figure(
                    data=[go.Scatter(x=dataframe[somedata1].values.tolist(),
                                    y=dataframe[somedata2].values.tolist(),
                                    mode = 'markers',
                                    marker = dict(
                                        size=5,
                                        color='rgb(0,0,153)')
                                    )
                         ],
                    layout=go.Layout(
                        title=title,
                        margin=go.Margin(l=40, r=0, t=40, b=30)
                    )
                ),
                style={'height': 300},
            id='grapha2'
            )

    
def graph_scatter2(somedata1, somedata2, title, dataframe,gid):
        return dcc.Graph(
                figure=go.Figure(
                    data=[go.Scatter(x=dataframe[somedata1].values.tolist(),
                                    y=dataframe[somedata2].values.tolist(),
                                    mode = 'markers',
                                    marker = dict(
                                        size=5,
                                        color='rgb(0,0,153)')
                                    )
                         ],
                    layout=go.Layout(
                        title=title,
                        margin=go.Margin(l=40, r=0, t=40, b=30)
                    )
                ),
                style={'height': 300},
            id='graphb2'
            )

    
# def run_ttest(a):
#     return print(stats.ttest_ind(df[a].dropna(), df2[a].dropna()))



@app.callback(
    Output('graph1','children'),
    [Input('dropdown1','value')]
)
def generate_graph1(value1):
    if len(value1)==1:
        return graph_histogram1(value1[0], 'REAL DATA', df,1)
    elif len(value1)==2:
        return graph_scatter1(value1[0],value1[1], 'REAL DATA',df,1)
    else:
        return "Please select up to two features "

    
@app.callback(
    Output('graph2','children'),
    [Input('dropdown1','value')]
)
def generate_graph2(value1):
    if len(value1)==1:
        return graph_histogram2(value1[0], 'SYNTHETIC DATA', df2,1)
    elif len(value1)==2:
        return graph_scatter2(value1[0],value1[1], 'SYNTHETIC DATA',df2,1)
    else:
        return " "

@app.callback(
    Output('debugger','children'),
    [Input('dropdown1','value') ]
)
def display_message(value1):
    if len(value1)==1:
        return "comparing histograms of {0}".format(value1[0])
    elif len(value1)==2:
        return "comparing scatterplots of {0} (x-axis) and {1} (y-axis)".format(value1[0], value1[1])
    else:
        return " "
   
    
    
app.css.append_css({
    'external_url': 'https://codepen.io/chriddyp/pen/bWLwgP.css'
})

if __name__ == '__main__':
    app.run_server(debug=True)
```
